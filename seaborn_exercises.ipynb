4. Load the chipotle dataset from SQL, create a bar chart that shows the 4 most popular items and the revenue produced by each.

```python
db = 'chipotle'
engine = create_engine(env.get_db_url(db))

query = text('select * from orders')
chip_df = pd.read_sql(query, engine.connect())
chip_df

```

```python
chip_df.item_price.astype(str)
```

```python
chip_df['item_price'] = chip_df['item_price'].str.replace('$', '').astype(float)
```

```python
chip_df[['item_name','item_price']].value_counts()
```

```python
top_four = chip_df.groupby('item_name').quantity.sum().nlargest(4)
top_four
```

```python
all_prices = chip_df.groupby('item_name').item_price.sum()
merged_df = pd.merge(top_four, all_prices, on='item_name', how='inner')
merged_df = merged_df.reset_index()
merged_df
```

```python
#create the plot
sns.barplot(data= merged_df, y='item_name', x='item_price')

#add title and label
plt.title('Most Popular Items')
plt.xlabel('revenue generated')
plt.show()
```

5. Load the sleepstudy dataset from pydataset and read it's documentation. Use seaborn to create a line chart of all the individual subject's reaction times and a more prominant line showing the average change in reaction time.

```python
plt.figure(figsize=(12,6))
sns.lineplot(data=sleep, x='Days', y='Reaction', hue='Subject')
plt.show()
```

```python
sleep.Subject = sleep.Subject.astype(str)
```

```python
plt.figure(figsize=(12,8))

sns.lineplot(data=sleep, x='Days', y='Reaction', hue='Subject', alpha=.5)
sns.lineplot(data=sleep, x='Days', y='Reaction', 
             color = 'black', errorbar=('ci',False), linewidth=3)

plt.title('Reaction Time of Subjects Over Time')
plt.legend(loc='upper right')
plt.show()
```
