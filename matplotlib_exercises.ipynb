1. Use matplotlib to plot the following equation:
    y=x^2-x+2
 You'll need to write the code that generates the x and y points.
 Add an anotation for the point 0, 0, as the origin.

```python
x = range(1,100)
y = [n**2 - n + 2 for n in x]
```

```python
plt.plot(x,y)
plt.title('y = x^2 - x+2')
plt.xlabel('x axis')
plt.ylabel('y axis')

plt.show()
```

3.Combine the figures you created in the last step into one large figure with 4 subplots.

```python
x = range(1,100)

y1 = [math.sqrt(n) for n in x]
y2 = [(n**3) for n in x]
y3 = [2**n for n in x]
y4 = [(1 / (n+1)) for n in x]

plt.figure(figsize=(12,10))

plt.subplot(221)
plt.plot(x, y1, color='pink')
plt.title('y1')


plt.subplot(222)
plt.plot(x, y2, color='brown')
plt.title('y2=x^3')


plt.subplot(223)
plt.plot(x, y3, color='green')
plt.title('y3')


plt.subplot(224)
plt.plot(x, y4)
plt.title('y4=1/(x+1)')
plt.suptitle('Separate Charts')
plt.show()
```

4.Combine the figures you created in the last step into one figure where each of the 4 equations has a different color for the points. 
Be sure to include a legend and an appropriate title for the figure.

```python
x = range(1,100)

plt.title("All 4 Graphs Together")
plt.legend(loc='upper left')
plt.xlim(0,10)
plt.ylim(0,10)

plt.plot(x,y1, color='gray', label='y1')


plt.plot(x, y2, label='y2')


plt.plot(x, y3, color='purple', label='y3')
plt.xlabel('x axis')
plt.ylabel('y axis')


plt.plot(x, y4, label='y4')


plt.legend()
plt.show()
```
