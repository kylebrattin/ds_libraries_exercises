 Curves to graph
- y=0n+1 and label the curve "O(1)"
- y=log(n)and label the curve "O(log n)"
- y=n and label the curve "O(n)"
- y=n*log(n)and label it "O(n log n)"
- y=n^2 and label it "O(n^2)"
- y=2^n and label it "O(2^n)"
- y=n! and label it "O(n!)"
- y=n^n and label it "O(n^n)"

```python
x = list(range(1 , 20))

y1 = [((n*0) + 1) for n in x]
y2 = [math.log(n) for n in x]
y3 = [n for n in x]
y4 = [n * math.log(n) for n in x]
y5 = [n**2 for n in x]
y6 = [2**n for n in x]
y7 = [math.factorial(n) for n in x]
y8 = [n**n for n in x]

plt.plot(x, y1, label='O(1)')
plt.plot(x, y2, label='O(log n)')
plt.plot(x, y3, label='O(n)')
plt.plot(x, y4, label='O(nlogn)')
plt.plot(x, y5, label='n^2')
plt.plot(x, y6, label='2^n')
plt.plot(x, y7, label='n!')
plt.plot(x,y8, label='n^n')
plt.xlim(1,20)
plt.ylim(0,50)
plt.title('Big O Notation')
plt.xlabel('Elements')
plt.ylabel('Operations')
plt.legend()
plt.show()
```
